## Speed

```python
task_two=data
task_two['time_diff'] = task_two.groupby(['identifier'])['datetime'].diff()#calculate difference between time within each hurricane
```

```python
data
```

```python
def haversine(lon1, lat1, lon2, lat2, earth_radius=6367):
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
    dlon=lon2-lon1
    dlat=lat2-lat1
    a=np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
    c= 2*np.arcsin(np.sqrt(a))
    km=earth_radius*c
    return km
```

```python
locs = data[['latitude', 'longitude']]
locs.columns = ['lats', 'longs']
locs = locs.join(data.groupby('identifier')[['latitude', 'longitude']].shift())#creates locations dataframe
task_two['dist'] = locs.apply(lambda x: haversine(x.longitude, x.latitude, x.longs, x.lats), axis=1)#applies function to location dataframe and adds values to the main dataframe
```

```python
task_two['speed']=task_two.dist*3600/task_two.time_diff.dt.total_seconds()#convert distance and time to what we need
task_two
```

## Resampling and Visualization

### Joaquin vs Maria

```python
task_three=task_two[((task_two.name=='MARIA') & (task_two.identifier.str[4:]=='2017')) | ((task_two.name=='JOAQUIN') & (task_two.identifier.str[4:]=='2015'))].reset_index()#create new dataframe with only 2015 joaquin and 2017 maria
```

```python
task_three=task_three.drop(['time_diff'],axis=1)#drop time_diff because altair doesn't like it and we do not need it
```

```python
import altair as alt#set up altair
alt.renderers.enable('notebook')
alt.data_transformers.disable_max_rows()
from altair import datum
```

```python
task_threeM=task_three[task_three.name=="MARIA"]
task_threeJ=task_three[task_three.name=="JOAQUIN"]

#combining dataframes
canes=pd.concat([task_threeJ,task_threeM])
canes
```

```python
chart1 = alt.Chart(task_threeJ).mark_line(size=2).encode(#graphing joaquin's speed over time
    alt.X('datetime:T'),
    alt.Y('speed:Q')
).transform_filter(alt.datum.name == "JOAQUIN")
chart1
```

```python
chart2 = alt.Chart(task_threeM).mark_line(size=2).encode(#graphing maria's speed over time
    alt.X('datetime:T'),
    alt.Y('speed:Q')
).transform_filter(alt.datum.name == "MARIA")
chart2
```

### Superimposed Plots

```python
task_three_twoM = task_threeM.reset_index()
task_three_twoJ = task_threeJ.reset_index()
```

```python
task_three_twoJ.datetime=task_three_twoJ.datetime+(task_three_twoM.datetime[0]-task_three_twoJ.datetime[0])#do math so the hurricanes start at the same time
moreCanes = task_three_twoJ.join(task_three_twoM, how='outer',rsuffix="_1")
moreCanes=moreCanes.drop(['num_pts', 'record_id', 'status', 'index', 'level_0', 'num_pts_1', 'record_id_1', 'status_1', 'index_1'], axis=1)
moreCanes
```

```python
chart3 = alt.Chart(moreCanes).mark_line().encode(#graphing maria's speed over time, but shifted to the same start as joaquin
    alt.X('datetime_1:T'),
    alt.Y('speed_1:Q'),
    color='name_1:N'
).transform_filter(alt.datum.name_1 == "MARIA")

chart4 = alt.Chart(moreCanes).mark_line().encode(#graphing joaquin's speed over time
    alt.X('datetime:T'),
    alt.Y('speed:Q'),
    color='name:N'
).transform_filter(alt.datum.name == "JOAQUIN")
chart4

chart4+chart3#superimpose the plots, Joaquin cut short because Maria has no points to compare beyond graph
```

### Difference Plot

```python
moreCanes['diffs']=np.nan#calculate differences between hurricane speeds
for i in range(len(moreCanes.datetime)):
    for j in range(len(moreCanes.datetime_1)):
        if moreCanes.datetime[i]==moreCanes.datetime_1[j]:
            moreCanes['diffs'][j] = moreCanes.speed[i]-moreCanes.speed_1[j]
moreCanes
```

```python
def plot_canes(df):#create plotting function
    return alt.Chart(df).mark_line().encode(
        x="datetime:T",y="diffs:Q")
```

```python
plot_canes(moreCanes)
```

### Resampling

```python
#Resample and interpolate each hurricane individually
task_three_fM=task_three_twoM
task_three_fM=task_three_fM.set_index('datetime')
task_three_fM=task_three_fM.resample('180min').interpolate()
task_three_fM=task_three_fM.reset_index()

task_three_fJ=task_three_twoJ
task_three_fJ=task_three_fJ.set_index('datetime')
task_three_fJ=task_three_fJ.resample('180min').interpolate()
task_three_fJ=task_three_fJ.reset_index()
```

```python
#Recombine hurricanes and calculate differences
endCanes = task_three_fJ.join(task_three_fM, how='outer',rsuffix="_1")
endCanes=endCanes.drop(['num_pts', 'record_id', 'status', 'index', 'level_0', 'num_pts_1', 'record_id_1', 'status_1', 'index_1'], axis=1)

endCanes['diffs']=np.nan
for i in range(len(endCanes.datetime)):
    for j in range(len(endCanes.datetime_1)):
        if endCanes.datetime[i]==endCanes.datetime_1[j]:
            endCanes['diffs'][j] = endCanes.speed[i]-endCanes.speed_1[j]
endCanes
```

```python
plot_canes(endCanes)
```
